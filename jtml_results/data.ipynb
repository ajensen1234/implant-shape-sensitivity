```python
import numpy as np 
```

```python
# read the data as a np array
def read_jts(file_path):
    data = np.genfromtxt(file_path,skip_header=2, dtype=float, delimiter=',')
    if len(data[1,:]) == 7:
        return data[:,:-1]
    else:
        return data
def rmsd(vec1, vec2):
    # first, we need to get the sum of squares difference
    sum = np.sum([(x-y)**2 for x,y in zip(vec1,vec2)])
    return np.sqrt(sum / len(vec1))
```

```python
jtml = read_jts('./movement1_scapula_optimize_all.jtak')
print(jtml)
jts = read_jts('./sca.jts')
print(jts)
```

```python
x_t_diff = rmsd(jtml[:,0], jts[:,0])
y_t_diff = rmsd(jtml[:,1], jts[:,1])
z_t_diff = rmsd(jtml[:,2], jts[:,2])
z_r_diff = rmsd(jtml[:,3], jts[:,3])
x_r_diff = rmsd(jtml[:,4], jts[:,4])
y_r_diff = rmsd(jtml[:,5], jts[:,5])
vals = ["X Translation RMSD", "Y Translation RMSD", "Z Translation RMSD", "X Rotation RMSD", "Y_Rotation RMSD", "Z Rotation RMSD" ]
```

```python
for val_name, val in zip(vals,[x_t_diff, y_t_diff, z_t_diff, x_r_diff, y_r_diff, z_r_diff]):
    print(val_name)
    print(val)
    print("===================")
```
